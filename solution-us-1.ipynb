## Section 1
> Consider an ultrasound beam that hits the boundary between two materials perpendicularly. Using the information provided in the lecture slides, show that transmission and reflection at the boundary conserve total beam intensity.

*Text solution:*

Let $p_{inc}$, $p_{trans}$, and $p_{ref}$ denote the pressure amplitudes of the incident, transmitted, and reflected beams, respectively. According to the lecture (for perpendicular incidence), they are related by

$$p_{ref}=p_{inc}\frac{Z_2-Z_1}{Z_2+Z_1}$$

$$p_{trans}=p_{inc}\frac{2Z_2}{Z_2+Z_1}$$

The intensity of a beam is given by
    
$$I=\frac{1}{2}p\,u_z$$

where $u_z$ denotes particle velocity amplitude, which is equal to pressure divided by impedance, so
    
$$I=\frac{p^2}{2Z}$$

Now, letâ€™s see if the sum of intensities of the transmitted and reflected beams is equal to that of the incident beam:
    
$$\frac{p_{trans}^2}{2Z_2}+\frac{p_{ref}^2}{2Z_1}=\frac{p_{inc}^2}{2Z_1}\quad?$$
    
Substitute pressures:

$$\frac{4p_{inc}^2 Z_2^2}{Z_2(Z_2+Z_1)^2}+\frac{p_{inc}^2(Z_2-Z_1)^2}{Z_1(Z_2+Z_1)^2}=\frac{p_{inc}^2}{Z_1}\quad?$$

$$\frac{4Z_1 Z_2}{(Z_2+Z_1)^2}+\frac{(Z_2-Z_1)^2}{(Z_2+Z_1)^2}=1\quad?$$

$$\frac{Z_2^2+2Z_1 Z_2 +Z_1^2}{(Z_2+Z_1)^2}=1\quad!$$


By subtracting the first two equations, one can also obtain a simple relationship that is independent of wave impedances:

$$p_{trans}-p_{ref}=p_{inc}$$

Some of you were tempted to work from this relationship; however, conservation of intensity cannot be derived from this equation alone. To appreciate this, consider adding any $\Delta p$ to both $p_{trans}$ and $p_{ref}$. Then, the difference equation will still hold while the sum of the intensities of the transmitted and reflected beams can assume any value.

## Section 2

Consider a flat transducer with a circular face (diameter = 3 mm) performing 1D ultrasound imaging of two layers of tissue (fat and muscle) as shown below, using pulses of frequency 3 MHz and duration 1 $\mu$s.

### Task 2.1
> Where is the near-field boundary (NFB)?

*Text solution:*

$$\textrm{NFB}=\frac{r^2}{\lambda_{fat}}=\frac{r^2 f}{c_{fat}}=\frac{(0.0015\, \textrm{m})^2 \cdot 3 \cdot 10^6\, \textrm{Hz}}{1476\, \textrm{m/s}}=0.457\, \textrm{cm}$$

So the NFB is still within the fat layer; therefore, it is correct to consider only the wavelength in fat to calculate it.

### Task 2.2
> Let the maximum (i.e., on-axis) pulse intensity be 10 mW/cm$^2$ at the level of the NFB. What are the on-axis pressure and intensity levels entering (i.e., transmitted into) the muscle layer?
<br><br>Neglect attenuation resulting from beam broadening.

*Text solution:*

At the level of the NFB, the on-axis pressure amplitude is

$$p_{NFB}=\sqrt{2 I Z_{fat}}=\sqrt{2\cdot 100\,\frac{\textrm{W}}{\textrm{m}^2}\cdot 1.36\cdot 10^6\, \frac{\textrm{kg}}{\textrm{m}^2\textrm{s}}}=16.49\, \textrm{kPa}$$

Until it reaches the interface, the beam crosses another 0.543 cm of fat. The attenuation involved is

$$0.543\, \textrm{cm} \cdot 0.5\,\frac{\textrm{dB}}{\textrm{cm MHz}}\cdot 3\, \textrm{MHz}=0.81\,\textrm{dB}$$

The attenuated pressure at 1 cm depth is then

$$p_{1cm,fat}=16.49\, \textrm{kPa} \cdot 10^{\,-\frac{0.81\, \textrm{dB}}{20\, \textrm{dB}}}=15.02\,\textrm{kPa}$$

The transmitted beam has pressure

$$p_{trans}=p_{1cm,fat}\frac{2Z_{muscle}}{Z_{muscle}+Z_{fat}}=15.02\,\textrm{kPa}\cdot\frac{2\cdot1.66\cdot10^6\,\textrm{kg/m}^2\textrm{s}}{1.66\cdot10^6\,\textrm{kg/m}^2\textrm{s}+1.36\cdot10^6\,\textrm{kg/m}^2\textrm{s}}=16.51\,\textrm{kPa}$$

and hence intensity

$$I_{trans}=\frac{p_{trans}^2}{2Z_{muscle}}=\frac{\left(16.51\cdot10^3\,\textrm{Pa}\right)^2}{2\cdot1.66\cdot10^6\,\textrm{kg/m}^2\textrm{s}}=82.1\,\frac{\textrm{W}}{\textrm{m}^2}=8.21\,\frac{\textrm{mW}}{\textrm{cm}^2}$$

### Task 2.3
> What is the peak particle velocity upon entrance of the pulse into the muscle layer?

*Text solution:*

$$u_{z,trans}=\frac{p_{trans}}{Z_{muscle}}=\frac{16.51\cdot10^3\,\textrm{Pa}}{1.66\cdot10^6\,\textrm{kg/m}^2\textrm{s}}=9.95\, \frac{\textrm{mm}}{\textrm{s}}$$

### Task 2.4
> At which depth in the muscle has the sound intensity dropped to 1 mW/cm$^2$?
<br><br>Assume 2 dB/cm/MHz of attenuation in muscle.

*Text solution:*

At intensity 1 mW/cm$^2$, the sound pressure is

$$p=\sqrt{2I Z_{muscle}}=\sqrt{2\cdot 10\,\frac{\textrm{W}}{\textrm{m}^2}\cdot1.66\cdot10^6\,\frac{\textrm{kg}}{\textrm{m}^2\textrm{s}}}=5.76\, \textrm{kPa}$$

The corresponding attenuation in dB is

$$20\, \textrm{log}_{10}\left( \frac{5.76\, \textrm{kPa}}{16.51\, \textrm{kPa}}\right)\textrm{dB}=-9.14\, \textrm{dB}$$

which corresponds to

$$\Delta z=\frac{-9.14\, \textrm{dB cm MHz}}{2\,\textrm{dB}\cdot 3\,\textrm{MHz}}=1.52\, \textrm{cm}$$

So the beam is down to 1 mW/cm$^2$ at a total depth of 2.52 cm.

### Task 2.5
> What is the lateral resolution at that depth?
<br><br>Assume a negligible difference in wavelength between the fat and muscle layers.

*Text solution:*

The broadening angle, given by that of muscle because of the greater depth passed through in the muscle layer, is

$$\theta=\textrm{arcsin}\left(\frac{0.61\,\lambda_{muscle}}{r}\right)=\textrm{arcsin}\left(\frac{0.61\,c_{muscle}}{r\,f}\right)=\textrm{arcsin}\left(\frac{0.61\cdot 1568\,\textrm{m/s}}{0.0015\,\textrm{m}\cdot3\cdot10^6\,\textrm{Hz}}\right)=12.3^{\circ}$$

Starting at a width of 3 mm at the level of the NFB, the beams broadens to

$$\Delta x=0.3\,\textrm{cm}+2\cdot(2.52\,\textrm{cm}-0.457\,\textrm{cm})\,\textrm{tan}(12.3^{\circ})=1.20\,\textrm{cm}$$

upon reaching a depth of 2.52 cm.



### Task 2.6
> What is the axial resolution in fat and muscle?

*Text solution:*

$$\Delta z_{fat}=\frac{p_d\,c_{fat}}{2}=\frac{1\cdot10^{-6}\,\textrm{s}\cdot 1476\,\textrm{m/s}}{2}=0.738\,\textrm{mm}$$
$$\Delta z_{muscle}=\frac{p_d\,c_{muscle}}{2}=\frac{1\cdot10^{-6}\,\textrm{s}\cdot 1568\,\textrm{m/s}}{2}=0.784\,\textrm{mm}$$
