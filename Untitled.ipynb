```python
import numpy as np
import matplotlib.pyplot as plt

# 参数设置
A = 1          # 脉冲幅度
T = 2          # 周期
tau = 1        # 脉冲宽度
sample_rate = 1000  # 采样率
t = np.linspace(0, 4 * T, sample_rate * 4 * T)  # 生成时间序列

# 生成矩形脉冲波
pulse = np.where((t % T) < tau, A, 0)

# 傅里叶级数计算
n_terms = 10  # 取傅里叶级数的项数
F = np.zeros_like(t)

for n in range(1, n_terms + 1):
    # 傅里叶系数计算
    coeff = 2 * A / (n * np.pi) * np.sin(n * np.pi * tau / T)
    F += coeff * np.cos(n * 2 * np.pi * t / T)

# 添加直流分量
F += A * tau / T

# 绘图
plt.figure(figsize=(12, 8))

# 绘制脉冲波
plt.subplot(2, 1, 1)
plt.plot(t, pulse, label='矩形脉冲波', color='blue')
plt.title('矩形脉冲波')
plt.xlabel('时间 (s)')
plt.ylabel('幅度')
plt.xlim(0, 4 * T)  # 设定x轴范围
plt.ylim(-0.5, 1.5)  # 设定y轴范围
plt.grid()
plt.legend()

# 绘制傅里叶级数的近似波形
plt.subplot(2, 1, 2)
plt.plot(t, F, label='傅里叶级数近似', color='red')
plt.title('傅里叶级数近似')
plt.xlabel('时间 (s)')
plt.ylabel('幅度')
plt.xlim(0, 4 * T)  # 设定x轴范围
plt.ylim(-0.5, 1.5)  # 设定y轴范围
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
